# Step 5 : Modeling Algoritma

```python
# Fungsi double exponential smoothing
def double_exponential_smoothing(data, alpha, beta):
    smoothed = [data[0]]
    trend = [data[1] - data[0]]
    
    for i in range(1, len(data)):
        smoothed.append(alpha * data[i] + (1 - alpha) * (smoothed[i-1] + trend[i-1]))
        trend.append(beta * (smoothed[i] - smoothed[i-1]) + (1 - beta) * trend[i-1])
    
    return smoothed
```

```python
# Fungsi double exponential smoothing forecast
def double_exponential_smoothing_forecast(data, alpha, beta, steps):
    smoothed = double_exponential_smoothing(data, alpha, beta)
    last_smoothed = smoothed[-1]
    last_trend = smoothed[-1] - smoothed[-2]
    forecast = []
    
    for i in range(steps):
        forecast.append(last_smoothed + (i + 1) * last_trend)
    
    return forecast
```

```python
# Pilih kolom Jumlah Populasi
jumlah_populasi = df['Jumlah'].values

# Parameter alpha dan beta (faktor penghalusan)
alpha = 0.2
beta = 0.2

# Lakukan double exponential smoothing
smoothed_data = double_exponential_smoothing(jumlah_populasi, alpha, beta)

# Buat data frame hasil smoothing
hasil_smoothing = pd.DataFrame({'Tahun': df['Tahun'], 'Jumlah': jumlah_populasi, 'Smoothed Jumlah Populasi': smoothed_data})


# Melakukan forecasting untuk 5 tahun kedepan
tahun_terakhir = int(df['Tahun'].iloc[-1])
for i in range(1, 5):
    tahun = str(tahun_terakhir + i)
    prediksi = double_exponential_smoothing_forecast(smoothed_data, alpha, beta, i)
    smoothed_data = np.append(smoothed_data, prediksi)
    hasil_smoothing.loc[len(hasil_smoothing)] = [tahun, None, prediksi[0]]


# Cetak data frame hasil smoothing dan prediksi
print(hasil_smoothing)
```

```python
# Visualisasi data asli, hasil smoothing, dan prediksi
plt.figure(figsize=(10, 6))
plt.plot(df['Tahun'], jumlah_populasi, label='Populasi Domba di Indonesia')
plt.plot(hasil_smoothing['Tahun'], hasil_smoothing['Smoothed Jumlah Populasi'], label='Prediksi')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
# Lakukan double exponential smoothing
smoothed_data = double_exponential_smoothing(jumlah_populasi, alpha, beta)

# Buat data frame hasil smoothing
hasil_smoothing = pd.DataFrame({'Tahun': df['Tahun'], 'Jumlah': jumlah_populasi, 'Smoothed Jumlah Populasi': smoothed_data})

# Evaluasi model menggunakan Mean Absolute Error (MAE)
mae = mean_absolute_error(jumlah_populasi, smoothed_data)
print("Mean Absolute Error (MAE):", mae)

# Evaluasi model menggunakan Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(jumlah_populasi, smoothed_data)
print("Mean Absolute Percentage Error (MAPE):", mape, "%")
```
